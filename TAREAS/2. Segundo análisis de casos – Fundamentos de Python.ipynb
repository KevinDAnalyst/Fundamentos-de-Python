# Problema #1: Calculadora de propina inteligente

## Problema

La propina debe calcularse dinamicamente sobre cualquier monto de la cuenta y cualquier porcentaje de propina. Basado en esto debe dar la propina y el total final.

## Entrada

- Total de la cuenta (Número decimal).
- Porcentaje de propina deseado (Número decimal).

## Proceso

- Limpiar las entradas eliminando espacios y símbolos como $ o %.
- Validar que los datos ingresados sean valores numéricos correctos.
- Dividir el porcentaje entre 100 para trabajar con decimales.
- Multiplicar el total de la cuenta por el porcentaje de propina para encontrar el monto de la propina.
- Sumar el monto de la propina al total original de la cuenta.
- Verificar que ninguno de los valores ingresados sea menor a cero.

## Salida esperada

- Un mensaje que muestra la propina
- Un mensaje que muestra el total final a pagar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | TC: 13500 PP: 5% | Propina: ₡675. - Total Final: ₡14,175 | Propina: ₡675 - Total Final: ₡14,175 |
| Caso límite | TC: 23250 PP: 0% | Propina: ₡0. - Total Final: ₡23,250 | Propina: ₡0 - Total Final: ₡23,250 |
| Caso inválido | TC: -12480 PP: 10% | Error. No se admiten valores inferiores a 0 | Error. No se admiten valores inferiores a 0 |


```python
# Escriba la solución aquí

try:
    # Entrada
    total_cuenta = float(input("Total de la cuenta: ").replace('$', '').strip())
    porcentaje_propina = float(input("Porcentaje de propina: ").replace('%', '').strip()) / 100

    # Proceso
    monto_propina = total_cuenta * porcentaje_propina
    total_final = total_cuenta + monto_propina

    # Salida
    if porcentaje_propina < 0 or total_cuenta < 0:
        print('Error. No se admiten valores inferiores a 0')
    else:
        print(f"Propina: ₡{monto_propina:.0f} - Total final: ₡{total_final:,.0f}")

except Exception as e:
    print("Error. Has ingresado un dato invalido")
```

# Problema #2: Detector de número par/impar

## Problema

Se debe poder ingresar cualquier entero hacia el infinito y decir si es par o impar.

## Entrada

- Ingresar un entero

## Proceso

- Validar que la entrada sea exclusivamente un número entero en notación numérica.
- Calcular el residuo de la división del número entre dos mediante el operador de módulo (%).
- Evaluar si el residuo es igual a cero para determinar que el número es par.
- Definir el número como impar en caso de que el residuo sea diferente de cero.

## Salida esperada

- Mensaje en pantalla que especifica si el número es par o impar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 20 | El numero 20 es par | El numero 20 es par |
| Caso límite | 1471 | El numero 13,471,237 es impar | El numero 13,471,237 es impar |
| Caso inválido | 4.5 | Error. Solo se admiten numeros enteros en notacion numerica | Error. Solo se admiten numeros enteros en notacion numerica |


```python
# Escriba la solución aquí
try:
    numero_entero = int(input("Ingrese un numero entero: ").strip())
    
    if numero_entero % 2 == 0:
        print(f"El numero {numero_entero:,} es par")
    else:
        print(f"El numero {numero_entero:,} es impar")

except ValueError as e:
    print(f"Error. Solo se admiten numeros enteros en notacion numerica")
```

# Problema #3: Clasificador de edad

## Problema

Se debe ingresar una edad y poder clasificar a la persona entre menor, adulto o adulto mayor

## Entrada

- Edad

## Proceso

- Validar que el dato ingresado corresponda a un número entero.
- Verificar si la edad se encuentra en el rango de 0 a 17 años para clasificar como menor de edad.
- Evaluar si la edad está entre 18 y 64 años para definir la categoría de adulto.
- Comprobar si la edad se ubica entre 66 y 120 años para asignar la condición de adulto mayor.
- Clasificar como fuera de rango cualquier valor numérico que no encaje en los límites anteriores.

## Salida esperada

- Mensaje en pantalla que indica la clasificación de edad correspondiente o una advertencia de error.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 25 | Usted es un Adulto | Usted es un Adulto |
| Caso límite | 0 | Usted es menor de edad | Usted es menor de edad |
| Caso inválido | -14 | Usted ha ingresado una edad fuera de rango | Usted ha ingresado una edad fuera de rango |


```python
# Escriba la solución aquí
try:
    edad = int(input("Ingrese su edad: "))

    if edad >= 0 and edad < 18:
        print("Usted es menor de edad")
    elif edad < 65 and edad >= 0:
        print("Usted es un adulto")
    elif edad > 65 and edad <= 120:
        print("Usted es un Adulto Mayor")
    else:
        print("Usted ha ingresado una edad fuera de rango")
except ValueError:
    print("Debe ingresar un numero entero")

```

# Problema #4: Validador de compra con descuento

## Problema

ECalcular el monto final a pagar por un cliente aplicando un descuento del 10% únicamente si es considerado cliente frecuente.

## Entrada

- Nombre del cliente (Texto).
- Monto de la compra (Número decimal).
- Condición de cliente frecuente (Texto: "Si" o "No").

## Proceso

- Validar que el monto ingresado sea un número válido.
- Evaluar si la respuesta sobre la frecuencia del cliente es afirmativa ("si").
- Restar el 10% al monto original si es cliente frecuente.
- Mantener el monto original intacto si no cumple la condición.

## Salida esperada

- Mensaje personalizado con el nombre del cliente y el total definitivo a pagar en colones.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | Kevin, 10000, no | Kevin, su total a pagar es de ₡10,000.00 | Kevin, su total a pagar es de ₡10,000.00 |
| Caso límite | Kevin, 0.05, si | Kevin, por ser un cliente frecuente se le aplicara un descuento del 10%. Su total a pagar es de ₡0.04 | Kevin, por ser un cliente frecuente se le aplicara un descuento del 10%. Su total a pagar es de ₡0.04 |
| Caso inválido | Kevin, cinco mil | El numero ingresado no es valido | El numero ingresado no es valido |


```python
# Solicita cliente, monto y si es cliente frecuente. Aplica descuento según reglas.
try:
    cliente = input("Ingrese el nombre del cliente: ").strip().capitalize()
    monto_compra = float(input("Monto de compra: "))
    cliente_frecuente = input("ES un cliente frecuente? (Si/No)").lower().strip()

    if cliente_frecuente == 'si':
        monto_final = monto_compra - (monto_compra * 0.10)
        print(f"{cliente}, por ser un cliente frecuente se le aplicara un descuento del 10%\nSu total a pagar es de ₡{monto_final:,.2f}")
    else:
        monto_final = monto_compra
        print(f"{cliente}, su total a pagar es de ₡{monto_final:,.2f}")

except ValueError:
    print("El numero ingresado no es valido.")


```
